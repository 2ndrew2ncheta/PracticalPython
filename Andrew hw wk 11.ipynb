```python
all_data=[]
for pdf in pdfs:
    print(f"now reading {pdf}")
    table=tabula.read_pdf(pdf, pages="all")
    df=pd.concat(table, ignore_index=True)
    all_data.append(df)
```

```python
BigDF=pd.concat(all_data, ignore_index=True)
BigDF
```

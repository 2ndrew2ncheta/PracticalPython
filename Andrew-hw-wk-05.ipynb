# Looted Benin Art Work Distribution

Scrape <a href="https://digitalbenin.org/Digital">the Benin site</a> to create a dataframe that contains the following scraped information about each institution:

- Museum name
- Museum address
- Number of disputed items

Export as a ```disputed-benin-artwork.csv```


````python

import pandas as pd
import requests
from random import randrange
import time
from bs4 import BeautifulSoup
````

````python
base_url="https://digitalbenin.org/institutions/"
````

````python
response=requests.get(base_url)
````

````python
soup=BeautifulSoup(response.text, "html.parser")
````

````python
print(soup.prettify)
````

````python
Museums=soup.find_all("div", class_="col-12 col-md-5")
Museums
````

````python
Just_museums=[inst.get_text() for inst in Museums]
Just_museums
````

````python
Num_objects=soup.find_all("div", class_="col-auto col-md-auto inst-col-count")
Num_objects
````

````python
Just_numbers=[number.get_text() for number in Num_objects]
Just_numbers
````

````python
Countries=soup.find_all("span", class_="text-truncate nav-link filter_main badge text-bg-dark")
Countries
````

````python
Just_Countries=[country.get_text() for country in Countries]
Just_Countries
````

````python
Table=list(zip(Just_museums, Just_Countries, Just_numbers))
Table
````

````python
df=pd.DataFrame(Table)
df
````

````python
df.columns=["Museum","Country","Number of Objects"]
df
````

````python
df.to_csv("disputed_benin_artwork.csv", index=False)
````
